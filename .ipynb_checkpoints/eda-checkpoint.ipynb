```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
columns = ['variance', 'skewness', 'kurtosis', 'entropy', 'class']
df = pd.read_csv('data.csv', names=columns)
```

```python
print("First rows: ")
df.head()
```

```python
print("Data set info: ")
df.info()
```

```python
print("Statistics: ")
print(df.describe())
```

```python
print("Number of banknotes in each class: ")
print(df['class'].value_counts())
```

```python
print("Mean values for each class: ")
comparison = df.groupby('class').mean()
print(comparison)
```

```python
print("Difference between classes: ")
print(comparison.loc[0] - comparison.loc[1])
```

```python
features = ['variance', 'skewness', 'kurtosis', 'entropy']

plt.figure(figsize=(10, 5))

for i, col in enumerate(features):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(x='class', y=col, data=df)
    plt.title(f'Feature distribution: {col}')

plt.tight_layout()
plt.show()
```

```python
correlation_matrix = df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='RdYlGn', fmt=".2f")
plt.title('Correlation matrix of banknote features')
plt.show()
```

```python
sns.pairplot(df, hue='class', palette='husl', markers=["o", "s"])
plt.suptitle('Relationships between all features', y=1.02)
plt.show()
```

# Banknote Authentication - EDA Report

## 1. Dataset Characteristics

* **Number of records:** 1,372 complete rows.
* **Features:** 4 mathematical image parameters (variance, skewness, kurtosis, entropy).
* **Class Balance:** The dataset is well-balanced  762 authentic banknotes (55%) and 610 counterfeit banknotes (45%).
* **Data Quality:** No missing values (NaN); all data points are numerical (float64/int64).

---

## 2. Key Parameters and Their Significance

* **Variance:** The most important differentiating feature. It shows a strong negative correlation with the target class (-0.72). Authentic banknotes exhibit significantly higher variance values than counterfeit ones.
* **Skewness:** An essential auxiliary parameter; it shows noticeable differences in distributions between the two classes.
* **Kurtosis:** This feature helps identify "outliers" occurring mainly in counterfeit banknotes, suggesting errors or inconsistencies in their printing technique.
* **Entropy:** The feature with the lowest diagnostic significance. Its correlation with the class is only -0.02, and the distributions for both groups almost entirely overlap.

---

## 3. Visualization Insights (Pairplot)



* Scatter plots reveal clear separation between the classes.
* Data points representing authentic and counterfeit banknotes form distinct clusters, suggesting that the model will be able to achieve very high classification accuracy.

---

## 4. Modeling Recommendations

* **Feature Scaling:** It is highly recommended to apply feature scaling (e.g., StandardScaler) before training the model due to the varying ranges of individual parameters.
* **Feature Selection:** The training process should prioritize **variance** and **skewness**, as they provide the most information gain for classification.
